#### RESIZING IMAGES

```python

#Interpolation method for resizing

image = cv2.imread('cvdata/im1.jpg')

half_sized = cv2.resize(image, (0,0),fx=0.5 , fy=0.5)
double_sized = cv2.resize(image, (502,402))

interpolate_image1 = cv2.resize(image ,(400,300), interpolation = cv2.INTER_NEAREST)

images = [image , half_sized , double_sized , interpolate_image1]
titles = [1,2,3,4]

for i in range(4):
    plt.subplot(2,2,i+1)
    plt.title(titles[i])
    plt.imshow(images[i])
plt.show()

print('Here new interpolation starts of type <INTER_AREA>')

interpolate_image2 = cv2.resize(image ,(400,300), interpolation = cv2.INTER_AREA)
images = [image , half_sized , double_sized , interpolate_image2]
titles = [1,2,3,4]

for i in range(4):
    plt.subplot(2,2,i+1)
    plt.title(titles[i])
    plt.imshow(images[i])
plt.show()
```

#### ERODING IMAGES

```python
import numpy as np
```

```python
#Syntax: cv2.erode(src, kernel[, dst[, anchor[, iterations[, borderType[, borderValue]]]]])


image = cv2.imread('cvdata/im2.jpg',0)

kernel = np.ones((9,9), np.uint8)

eroded_image = cv2.erode(image,kernel)
cv2.imshow('eroded', eroded_image)
cv2.waitKey(0)
cv2.destroyAllWindows()

#Here is an another example with BorderType parameter

image2 = cv2.imread('cvdata/erode.png')

kernel = np.ones((9,9), np.uint8)

eroded_image = cv2.erode(image2,kernel, cv2.BORDER_REFLECT)
cv2.imshow('eroded', eroded_image)
cv2.waitKey(0)
cv2.destroyAllWindows()

```

#### Blurring a image
##### Gaussian Blur

```python
#Gaussian Blur


#src  A Mat object representing the source (input image) for this operation.
#dst  A Mat object representing the destination (output image) for this operation.
#ksize  A Size object representing the size of the kernel.
#sigmaX  A variable of the type double representing the Gaussian kernel standard deviation in X direction.


#Median Blur


#src- It represents the source (input image).
#dst - It represents the destination (output image).
#ksize - It represents the size of the kernel.


#Bilateral Filter


#src - Source 8-bit or floating-point, 1-channel or 3-channel image.
#dst - Destination image of the same size and type as src.
#d - Diameter of each pixel neighborhood that is used during filtering. If it is non-positive, it is computed from sigmaSpace.
#sigmaColor - Filter sigma in the color space. A larger value of the parameter means that farther colors within the pixel neighborhood will be mixed together, resulting in larger areas of semi-equal color.
#sigmaSpace - Filter sigma in the coordinate space. A larger value of the parameter means that farther pixels will influence each other as long as their colors are close enough. When d>0 , it specifies the neighborhood size regardless of sigmaSpace . Otherwise, d is proportional to sigmaSpace.
```

```python

image = cv2.imread('cvdata/blur2.jpg')

#Gaussian Blur

Gaussian_blur = cv2.GaussianBlur(image, (3,3),3)

#Median Blur

median_Blur = cv2.medianBlur(image, 9)

#Bilateral Filter
bilateral_Filter = cv2.bilateralFilter(image, 16,200,200)

images = [image , Gaussian_blur , median_Blur , bilateral_Filter]

plt.figure(figsize=(27, 27))

for i in range(4):
    
    plt.subplot(1,4,i+1)
    plt.imshow(images[i])
plt.show()

#checking whether the blur has worked or not

org_px = image[100,180,0]
gauss_px = Gaussian_blur[100,180,0]
med_px = median_Blur[100,180,0]
bilateral_px = bilateral_Filter[100,180,0]

pixels = [org_px, gauss_px, med_px, bilateral_px]
pixel_names = ['original', 'gaussian', 'median', 'bilateral']
for i in range(4):
    print("pixel value for", pixel_names[i],"is", pixels[i])
    
while True:
    for i in range(3):
        if pixels[i] ==  pixels[i+1] :
            i+=1
        print("\n pixels are : ",pixels[i],",", pixels[i+1])
    
    print("\n Perfect Blur\n")
    print("selection of parameters for blurring images in great")
    break
```

##### Making Border around the image

```python
image = cv2.imread('cvdata/blur2.jpg')

border1 = cv2.copyMakeBorder(image,50,50,50,50,cv2.BORDER_REPLICATE )
border2 = cv2.copyMakeBorder(image,50,50,50,50,cv2.BORDER_CONSTANT , 210)
border3 = cv2.copyMakeBorder(image,50,50,50,50,cv2.BORDER_ISOLATED )
border4 = cv2.copyMakeBorder(image,50,50,50,50,cv2.BORDER_REFLECT101 )
border5 = cv2.copyMakeBorder(image,50,50,50,50,cv2.BORDER_WRAP )
border6 = cv2.copyMakeBorder(image,50,50,50,50,cv2.BORDER_REFLECT)

borders = [border1, border2, border3, border4, border5, border6]

plt.figure(figsize=(30,30))
for i in range(6):
    plt.subplot(2,3,i+1)
    plt.imshow(borders[i])
    
plt.show()
```

##### Grayscaling of images

```python
# Method 1

image = cv2.imread('cvdata/gray.png')
cv2.imshow('grayed', image)
cv2.waitKey(0)


gray_img = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2.imshow('grayed1', gray_img)
cv2.waitKey(0)


# Method 2

image1 = cv2.imread('cvdata/gray.png',0)
cv2.imshow('grayed2', image1)
cv2.waitKey(0)


# Method 3

image2 = cv2.imread('cvdata/gray.png')
row, col = image2.shape[:2]

for i in range(row):
    for j in range(col):
        image2[i,j] = sum(image2[i,j]) * 0.33
        
cv2.imshow("grayed3", image2)
cv2.waitKey(0)
cv2.destroyAllWindows()


#Plotting these figures all together (the color will be changed as the subplot has its own parameters)


figures = [image,gray_img, image1, image2]

plt.figure(figsize=(10,10))

for i in range(4):
    plt.subplot(2,2,i+1)
    plt.imshow(figures[i])
plt.show()
```
